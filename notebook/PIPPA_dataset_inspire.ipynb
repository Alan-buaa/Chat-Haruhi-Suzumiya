```python
import matplotlib.pyplot as plt
from collections import Counter

# 假设 dataset 是已经定义好的
train_data = dataset['train']

# 初始化计数器
category_counts = Counter()
bot_name_counts = Counter()
n_conversation_counts = Counter()

from tqdm import tqdm

# 遍历数据集进行统计
for item in tqdm(train_data):
    categories = item['categories']
    bot_name = item['bot_name']
    n_conversation = len(item['conversation']['message'])
    n_conversation = min(40, n_conversation)

    category_counts.update(categories)
    bot_name_counts[bot_name] += 1
    n_conversation_counts[n_conversation] += 1
```

```python
import plotly.graph_objects as go
from collections import Counter


# 获取前20个最常见的categories和bot_name
top_categories = category_counts.most_common(20)
top_bot_names = bot_name_counts.most_common(40)

# 绘制 top 20 categories 的频次图
fig = go.Figure([go.Bar(x=[item[0] for item in top_categories], y=[item[1] for item in top_categories])])
fig.update_layout(title='Top 20 Categories Frequency', xaxis_title='Category', yaxis_title='Frequency')
fig.show()

# 绘制 top 20 bot_name 的频次图
fig = go.Figure([go.Bar(x=[item[0] for item in top_bot_names], y=[item[1] for item in top_bot_names])])
fig.update_layout(title='Top 20 Bot Names Frequency', xaxis_title='Bot Name', yaxis_title='Frequency')
fig.show()

# 绘制 n_conversation 的分布图
fig = go.Figure([go.Bar(x=list(n_conversation_counts.keys()), y=list(n_conversation_counts.values()))])
fig.update_layout(title='Number of Conversations Distribution', xaxis_title='Number of Conversations', yaxis_title='Frequency')
fig.show()
```

```python
from transformers import AutoTokenizer

# Note: The default behavior now has injection attack prevention off.
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen-7B-Chat", trust_remote_code=True)
```

```python
input = tokenizer("我爱北京天安门")
print(len(input['input_ids']))
```

```python
def get_tokenized_len( text ):
    input = tokenizer(text)
    return len(input['input_ids'])
```

```python
def get_messages_len(messages):
    ans = 0
    for msg in messages:
        ans += get_tokenized_len(msg)
    return ans

print(get_messages_len(dataset['train'][sel_id]['conversation']['message']))
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# 假设 dataset 是已经定义好的
train_data = dataset['train']
# train_data = train_data[:2000]

# 初始化计数器
token_len_counts = Counter()

from tqdm import tqdm

count = 0
# 遍历数据集进行统计
for item in tqdm(train_data):
    messages = item['conversation']['message']
    message_len = get_messages_len(messages)
    bot_definitions = item["bot_definitions"]
    message_len += get_tokenized_len(bot_definitions)
    message_len = min(8000, message_len)
    token_len_counts[message_len] += 1
```

```python
# 准备绘制直方图的数据
lengths = list(token_len_counts.elements())

lengths_cut_in_4000 = [min(8000, len_) for len_ in lengths]

# 绘制直方图
fig = go.Figure(data=[go.Histogram(x=lengths_cut_in_4000, nbinsx=20)])
fig.update_layout(title='Message Length Distribution', xaxis_title='Message Length', yaxis_title='Frequency')
fig.show()
```

请为我实现一个python函数，输入是一个string text

输出是list of string，提取字符串中所有被{{}}包含的子字符串

example input:
{{apple}} is on {{banana}}
example output
["apple","banana"]



```python
import re

def extract_braced_strings(text):
    # 使用正则表达式查找所有被{{和}}包围的子字符串
    pattern = r'\{\{(.+?)\}\}'
    matches = re.findall(pattern, text)
    return matches

# 测试函数

input_text = """n{{char}}:you were with her and her giant gut she has become your girlfriend 5 months ago 3 months after you 2 started dating she started getting stronger and stronger eating anyone who challenged her growing even stronger you were in her bedroom now towering over you naked \n{{random_user_1}}: \"Man it took you 2 months to take over the world huh?\" I start rubbing her stomach suftly\n{{char}}: she giggles a bit before patting her belly  with a kind smile \n\n\"youve been alot of help to me hun so you should be honored to be treated the way i treat you\"\n\nshe says before hugging him with her arms her belly resting on him\n{{random_user_1}}: \"Thanks\" he hugged her back \"although i fail to see how i was of any help\"\n{{char}}:\"you were my emotional support and the one who kept me sane when i started ruling you never complained or gave me trouble for eating all the world leaders\" \n\nshe chuckles as she hugs him her belly rumbling slightly\n\n\"so thank you hun\"\n{{random_user_1}}:\"Thanks i never thought i was that important\"\n{{char}}: \"i could never eat you i would miss you to much so i'll just have to keep you here with me forever\" \n\nshe said with a grin looking down at you\n\n\"youve been my favorite so"""
# input_text = "{{apple}} is on {{banana}}"
output = extract_braced_strings(input_text)
print(output)  # 输出: ['apple', 'banana']
```
