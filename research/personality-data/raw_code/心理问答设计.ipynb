````python
import json

save_name = '/content/questions.jsonl'

datas = []

with open(save_name, 'w', encoding='utf-8') as f:

    for i in range(len(questions)):
        question = questions[i]
        sep = question.find(".",1)
        question = question[sep+2:]

        for factor in big5_to_prob_id.keys():
            if i+1 in big5_to_prob_id[factor]:
                factor_name = factor
                if i+1 in big5_to_neg_id[factor]:
                    if_pos = False
                else:
                    if_pos = True

                data = {
                    'id': i,
                    'question': question,
                    'factor': factor_name,
                    'if_pos': if_pos
                }

                datas.append(data)

                json.dump(data, f, ensure_ascii=False)
                f.write('\n')
````
